```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
import numpy as np
life_df=pd.read_csv("/content/Life expectancy.csv")
```

```python
life_df.describe()
```

```python
Q1 = life_df["Life expectancy"].quantile(0.25)
Q3 = life_df["Life expectancy"].quantile(0.75)
print(Q1,Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
print(lower_limit,upper_limit)
```

```python
life_df[(life_df["Life expectancy"] < lower_limit) | (life_df["Life expectancy"] > upper_limit)]
```
